```python
#Statistics Basic-1 Assignment
"""Q1. What is Statistics?"""
Ans: 
Statistics is the science of collecting, organizing and analyzing data.
                                    
                                    OR
    
Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, and
presentation of data. It is used in data science to analyze and interpret data to gain insights and 
make predictions. Statistics can be used to identify patterns in data, measure relationships between 
variables, and make predictions about future events.

"""Q2. Define the different types of statistics and give an example of when each type might be used."""

Ans: 
1. Descriptive Statistics: Descriptive statistics are used to summarize data and describe the characteristics
of a population. 
Examples include measures of central tendency (mean, median, mode) and measures of 
variability (standard deviation, range). Descriptive statistics can be used to describe the results of 
a survey or to compare different groups of data.

2. Inferential Statistics: Inferential statistics are used to make predictions or draw conclusions from 
a sample of data. 
Examples include hypothesis testing, correlation, and regression analysis. Inferential
statistics can be used to determine if there is a significant difference between two groups or to 
predict future outcomes.

"""Q3. What are the different types of data and how do they differ from each other? Provide an example of
each type of data."""

Ans: 
1.Ordinal Scale Data: Ordinal data is data that can be ordered or ranked. Examples include letter grades 
(A, B, C), survey responses (strongly agree, agree, disagree), and rankings (first, second, third).

2. Nominal Scale Data: Nominal data is a type of data that is used to label variables without providing
any quantitative value. Examples of nominal data include gender (male/female), religion (Christian, Muslim, 
Hindu, etc.), and political affiliation (Democrat, Republican, Independent).

3. Ratio Scale Data: Ratio data is a type of data that provides quantitative information and can be used
to compare values. Examples of ratio data include height, weight, age, and income.

4. Nominal Interval data is a type of data that has values that are ordered, but the intervals between 
the values are not equal. Examples of nominal interval data include temperature scales (e.g. Celsius, 
Fahrenheit, Kelvin), rating scales (e.g. 1-5 stars), and age ranges (e.g. 0-5, 6-10, 11-15).

"""Q4. Categorise the following datasets with respect to quantitative and qualitative data types:
(i) Grading in exam: A+, A, B+, B, C+, C, D, E
(ii) Colour of mangoes: yellow, green, orange, red
(iii) Height data of a class: [178.9, 179, 179.5, 176, 177.2, 178.3, 175.8,...]
(iv) Number of mangoes exported by a farm: [500, 600, 478, 672, ...]"""

Ans: 
(i) Quantitative 
(ii) Qualitative 
(iii) Quantitative 
(iv) Quantitative

"""Q5. Explain the concept of levels of measurement and give an example of a variable for each level."""
Ans: Levels of measurement refer to the way in which data is categorized and measured. There are four levels of measurement: nominal, ordinal, interval, and ratio. 

Nominal: A nominal level of measurement is used to classify data into categories that have no numerical
order. An example of a nominal variable is gender (male/female).

Ordinal: An ordinal level of measurement is used to classify data into categories that have an order or
ranking. An example of an ordinal variable is educational level (high school, college, graduate school).

Interval: An interval level of measurement is used to measure data that has an order and equal intervals
between values. An example of an interval variable is temperature (in degrees Celsius or Fahrenheit).

Ratio: A ratio level of measurement is used to measure data that has an order, equal intervals between 
values, and a true zero point. An example of a ratio variable is height (in inches or centimeters).

"""Q6. Why is it important to understand the level of measurement when analyzing data? Provide an
example to illustrate your answer."""
Ans: It is important to understand the level of measurement when analyzing data because different levels
of measurement require different types of statistical analysis. 
For example, if you are analyzing data that is measured on a nominal scale (e.g., gender), then you would 
use a chi-square test to determine if there is a significant difference between the two groups. On the 
other hand, if you are analyzing data that is measured on an interval or ratio scale (e.g., age), then 
you would use a t-test to determine if there is a significant difference between the two groups.

"""Q7. How nominal data type is different from ordinal data type."""
Ans: Nominal data type is a type of data that is used to label variables without providing any 
quantitative value. It is used to categorize data into mutually exclusive groups. 
Examples of nominal data include gender, nationality, religious affiliation, etc.

Ordinal data type is a type of data that is used to rank variables in order of magnitude or importance. 
It is used to assign a numerical value to a variable, such as 1st, 2nd, 3rd, etc. Examples of ordinal 
data include survey responses (e.g., strongly agree, agree, disagree), rankings (e.g., excellent, good,
average), etc.

"""Q8. Which type of plot can be used to display data in terms of range?"""
Ans: Box plot  

"""Q9. Describe the difference between descriptive and inferential statistics. Give an example of each
type of statistics and explain how they are used."""

Ans: Descriptive statistics are used to summarize and describe data. They are used to organize, present,
and describe data in a meaningful way. 
Examples of descriptive statistics include measures of central tendency (mean, median, mode), measures 
of variability (standard deviation, range), and frequency distributions. Descriptive statistics are used 
to describe the characteristics of a population or sample.

Inferential statistics are used to make predictions or inferences about a population based on a sample. 
Examples of inferential statistics include hypothesis testing, correlation, and regression. Inferential
statistics are used to draw conclusions from data and make predictions about a population.

"""Q10. What are some common measures of central tendency and variability used in statistics? Explain
how each measure can be used to describe a dataset."""

Ans: Common measures of central tendency include the mean, median, and mode. The mean is the average of 
all values in a dataset and is calculated by adding up all values and dividing by the number of values.
The median is the middle value in a dataset when all values are arranged in numerical order. The mode is the most frequently occurring value in a dataset.

Common measures of variability include the range, variance, and standard deviation. The range is the 
difference between the highest and lowest values in a dataset. The variance is a measure of how spread 
out the values are from the mean. The standard deviation is a measure of how much variation there is 
from the mean.
```
